```python
# load the packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python
# select reservoir name
reservoir_name = 'Dickson' # possible to replace by: Dickson, Bhumiboi, Waterton, Glen_canyon, Palisades, Trinity, Semione, Sirikit
#reservoir_name = "Bhumiboi"
#reservoir_name = "Sirikit"
#reservoir_name = 'Waterton'
```

```python
df_Inne = pd.read_csv('Hanasaki_Inne_'+reservoir_name+'.csv')
df_Inne ['timed'] = pd.to_datetime(df_Inne.timed)
df_Inne = df_Inne.set_index(['timed'])

df_Shervan = pd.read_csv('Hanasaki_Shervan_'+reservoir_name+'.csv')
df_Shervan ['timed']= pd.to_datetime(df_Shervan.timed)
df_Shervan = df_Shervan.set_index(['timed'])

if reservoir_name == "Sirikit": df_Shervan= df_Shervan['1990':'1994']; df_Inne= df_Inne['1990':'1994']
if reservoir_name == "Bhumiboi": df_Shervan= df_Shervan['1990':'1994']; df_Inne= df_Inne['1990':'1994']
```

### Timeseries outflow and storage

```python
import matplotlib
font = {'family' : 'Times New Roman',
         'weight' : 'bold',
         'size'   : 24}
matplotlib.rc('font', **font)

fig, axs = plt.subplots(1,1,figsize=(15,6))

plt.plot(df_Shervan.index, df_Shervan.Outflow_Observed, color="gray", label = "Outflow Observed")
plt.plot(df_Inne.index,    df_Inne.Outflow_Hanasaki,    label = "Outflow Hanasaki Inne")
plt.plot(df_Shervan.index, df_Shervan.Outflow_Hanasaki, label = "Outflow Hanasaki Shervan")


plt.legend()
plt.grid()
```

```python
import matplotlib
font = {'family' : 'Times New Roman',
         'weight' : 'bold',
         'size'   : 24}
matplotlib.rc('font', **font)

fig, axs = plt.subplots(1,1,figsize=(15,6))
plt.plot(df_Shervan.index, df_Shervan.Storage_Observed, color="gray", label = "Storage Observed")
plt.plot(df_Inne.index,    df_Inne.Storage_Hanasaki,    label = "Storage Hanasaki Inne")
plt.plot(df_Shervan.index, df_Shervan.Storage_Hanasaki, label = "Storage Hanasaki Shervan")

plt.legend()
plt.grid()
```

### Seasonal outflow and storage

```python
fig, axs = plt.subplots(1,1,figsize=(15,6))

plt.plot(df_Inne.Outflow_Hanasaki.groupby(df_Inne.index.dayofyear).mean(),    label = "Outflow Hanasaki Inne")
plt.plot(df_Shervan.Outflow_Hanasaki.groupby(df_Inne.index.dayofyear).mean(), label = "Outflow Hanasaki Shervan")
plt.plot(df_Shervan.Outflow_Observed.groupby(df_Inne.index.dayofyear).mean(), color="gray", label = "Outflow Observed")
plt.title('Seasonal outflow (m/s)', loc="right")
plt.legend()
plt.grid()
```

```python
fig, axs = plt.subplots(1,1,figsize=(15,6))

plt.plot(df_Inne.Storage_Hanasaki.groupby(df_Inne.index.dayofyear).mean(),    label = "Storage Hanasaki Inne")
plt.plot(df_Shervan.Storage_Hanasaki.groupby(df_Inne.index.dayofyear).mean(), label = "Storage Hanasaki Shervan")
plt.plot(df_Shervan.Storage_Observed.groupby(df_Inne.index.dayofyear).mean(), color="gray", label = "Storage Observed")
plt.title('Seasonal storage (m)', loc="right")
plt.legend()
plt.grid()
```
